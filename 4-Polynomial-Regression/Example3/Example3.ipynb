IMPORTING THE DATASET AND ASSIGNING X AND Y

```python
import pandas as pd
df = pd.read_csv("pyramid_scheme.csv")
x = df[["cost_price", "profit_markup","depth_of_tree", "sales_commission"]]
y = df[["profit"]]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

TRAIN THE MODEL

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2)
x_poly = poly_reg.fit_transform(x_train)

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_poly, y_train)
```

COMPARING WITH TRAINING SET

```python
y_pred = regressor.predict(x_poly)
print("TRAINING SET RESULTS")
from IPython.display import display_html
tab1 = x_train.head(20).style.set_table_attributes("style='display: inline'")
tab2 = y_train.head(20).style.set_table_attributes("style='display: inline'")
tab3 = pd.DataFrame(y_pred,columns=["Predicted Profit"]).head(20).style.set_table_attributes("style='display: inline'")
display_html(tab1._repr_html_()+"&nbsp;&nbsp;&nbsp;"+tab2._repr_html_()+"&nbsp;&nbsp;&nbsp;"+tab3._repr_html_(),raw=True)
```

```python
y_pred_t = regressor.predict(poly_reg.fit_transform(x_test))
print("TEST SET RESULTS")
tab1 = x_test.head(20).style.set_table_attributes("style='display: inline'")
tab2 = y_test.head(20).style.set_table_attributes("style='display: inline'")
tab3 = pd.DataFrame(y_pred_t,columns=["Predicted Profit"]).head(20).style.set_table_attributes("style='display: inline'")
display_html(tab1._repr_html_()+"&nbsp;&nbsp;&nbsp;"+tab2._repr_html_()+"&nbsp;&nbsp;&nbsp;"+tab3._repr_html_(),raw=True)
```
