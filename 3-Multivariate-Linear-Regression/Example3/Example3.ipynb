IMPORT THE DATASET AND ASSIGN X AND Y

```python
import pandas as pd
df = pd.read_csv("BMI.csv")
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values 
# Index : 0 - Extremely Weak 1 - Weak 2 - Normal 3 - Overweight 4 - Obesity 5 - Extreme Obesity
```

```python
x
```

```python
y[:5]
```

ENCODE CATEGORICAL DATA USING ONE LABEL ENCODER--GENDER

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[:,[0]]= le.fit_transform(x[:,[0]]).reshape(500,-1)
```

SPLIT THE SET INTO TRAINING AND TEST

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

TRAIN THE MODEL

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
y_pred=regressor.predict(x_train)
```

COMPARE THE PREDICTED RESULTS WITH TRAINING SET

```python
from IPython.display import display_html
tab1 = pd.DataFrame(x_train,columns=["Male?","Height","Weight"]).head(20).style.set_table_attributes("style='display: inline'")
tab2 = pd.DataFrame(y_train,columns=["BMI Index"]).head(20).style.set_table_attributes("style='display: inline'")
tab3 = pd.DataFrame(y_pred,columns=["Predicted BMI Index"]).head(20).style.set_table_attributes("style='display: inline'")
display_html(tab1._repr_html_()+"&nbsp;&nbsp;&nbsp;"+tab2._repr_html_()+"&nbsp;&nbsp;&nbsp;"+tab3._repr_html_(),raw=True)
```

COMPARE THE PREDICTED RESULTS WITH TEST SET

```python
y_pred_t=regressor.predict(x_test)
tab1 = pd.DataFrame(x_test,columns=["Male?","Height","Weight"]).head(20).style.set_table_attributes("style='display: inline'")
tab2 = pd.DataFrame(y_test,columns=["BMI Index"]).head(20).style.set_table_attributes("style='display: inline'")
tab3 = pd.DataFrame(y_pred_t,columns=["Predicted BMI Index"]).head(20).style.set_table_attributes("style='display: inline'")
display_html(tab1._repr_html_()+"&nbsp;&nbsp;&nbsp;"+tab2._repr_html_()+"&nbsp;&nbsp;&nbsp;"+tab3._repr_html_(),raw=True)
```
